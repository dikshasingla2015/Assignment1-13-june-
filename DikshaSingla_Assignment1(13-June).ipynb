```python
data_catamarca=data_df[['report_date','value']]
data_catamarca=data_catamarca.reset_index(drop=True)
data_catamarca['value']=pd.to_numeric(data_catamarca.value)
data_catamarca
```

# Month wise

```python
date=list(data_catamarca.report_date)
month=[]
for d in date:
    yr,mn,dy=d.split('-')
    month.append(mn)
data_catamarca['month']=month
data_catamarca
```

# week of the month wise

```python
import calendar, datetime
def week_of_month(date):
    cal_object = calendar.Calendar(6)
    month_calendar_dates = cal_object.itermonthdates(date.year,date.month)
    day_of_week = 1
    week_number = 1
    for day in month_calendar_dates:
        if day_of_week > 7:
            week_number += 1
            day_of_week = 1
        if date == day:
            break
        else:
            day_of_week += 1
    return week_number
```

```python
date=list(data_catamarca.report_date)
week=[]
for d in date:
    yr,mn,dy=d.split('-')
    date_formed=datetime.date(int(yr),int(mn),int(dy))
    week.append(str(week_of_month(date_formed)))
data_catamarca['week']=week
data_catamarca
```

```python
data_catamarca_week=data_catamarca.groupby(['month','week']).sum()
data_catamarca_week=data_catamarca_week.reset_index()
data_catamarca_week
```

```python
import calendar, datetime
def week_of_month(date):
    cal_object = calendar.Calendar(6)
    month_calendar_dates = cal_object.itermonthdates(date.year,date.month)
    day_of_week = 1
    week_number = 1
    for day in month_calendar_dates:
        if day_of_week > 7:
            week_number += 1
            day_of_week = 1
        if date == day:
            break
        else:
            day_of_week += 1
    return week_number
```

# Ques 4:Build a program to predict the correct word for wrongly typed word. for example whst => what, numbr =>number, good => good

```python
file=open('big.txt','r')
```

```python
text=file.read().lower()
```

```python
import string
table = str.maketrans('', '', string.punctuation)
text=text.translate(table)
```

```python
text=text.replace('\n',' ')
```

```python
text=text.replace('\t',' ')
```

```python
words=set(text.split(' '))
len(words)
```

```python
'' in words
```

```python
words.remove('')
```

```python
len(words)
```

```python
import difflib
import numpy as np
```

```python
difflib.get_close_matches('whst',words,n=10)
```

```python
def count_matched_characters(str1,str2):
    set_string1 = set(str1) 
    set_string2 = set(str2)
    matched_characters = set_string1 & set_string2
    return len(matched_characters),str2

def find_closest_match(input_str):
    maximum=0
    result=difflib.get_close_matches(input_str,words,n=10)
    print(result)
    if result[0]==input_str:
        print("closest match for {} is {}".format(input_str,result[0]))
    else:
        words_match_close=[]
        for wor in result:
            if (len(wor)==len(input_str)) or (len(wor)==(len(input_str)+1)):
                words_match_close.append(wor)
        print(words_match_close)
        for word in words_match_close:
            length,string=count_matched_characters(input_str,word)
            if length>maximum:
                maximum=length
                string_matched=string
        print("closest match for {} is {}".format(input_str,string_matched))

input_string=input("Enter a String: ")
print(len(input_string))
if len(input_string)<=0:
    print("Empty Input")
elif len(input_string)>0:
    if input_string.isalpha():
        find_closest_match(input_string)
    else:
        print("Word Contains Numeric Numbers")
```
